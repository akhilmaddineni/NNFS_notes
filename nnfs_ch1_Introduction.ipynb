Example of neural net : all dense layers  



Dense layers, the most common layers, consist of interconnected neurons. In a dense layer, each neuron of a given layer is connected to every neuron of the next layer, which means that its output value becomes an input for the next neurons

Each connection has weight associated with it ! and we can use it as a parameter to train the network . 


Weights and biases are tunable parameters in a network . 

Think of it like the equation : 

y = m*x + b 

Output = weight * input + bias 



As we can see weight as a slope in this linear equation and bias is an offset to the equation .


Step function : think of this like on-off switch 



you can represent this step fucntion as activation to neuron as  

output = activation(output)


A basic neural network contains Input and output layers and bunch of hidden layers  

For a classification problem we can decide on number of output layers we want to have . 

for classifying a image as cat or a dog we can have two output nodes . One for each class or only one node as an output to classify as cat or not cat (or vice versa)

The end goal for neural networks is to adjust their weights and biases (the parameters), so when applied to a yet-unseen example in the input, they produce the desired output.

overfitting : network kinda memorizes the training data and dont work well on test datasets . 

we divide the data which is not seen by network for validation . this is called "generalization" , a network should be able to predict the validation data which it has never seen somewhat accurately.
